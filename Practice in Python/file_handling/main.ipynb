close is import for handle buffer problem. If we do not close the file, it may cause memory leaks and other issues. It is always a good practice to close the file after we are done with it.

```python
file=open("shopping_list.txt",'r')
content=file.read()
file.close()
print(content)
```

good practice with open() as f: is used to automatically close the file after we are done with it. It is a good practice to use this method to handle files in Python.

```python
with open('fx.txt','r') as f:
    file=f.read()

print(file)
```

```python
with open('fx.txt','a') as f:
    file=f.write(sentence)


with open('fx.txt','r') as f:
    file=f.read()

print(file)
```
